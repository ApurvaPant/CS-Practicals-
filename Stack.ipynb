```python
stack = []

def isEmpty(stack):
    if len(stack) == 0:
        return True
    else:
        return False

def push(element,stack):
    stack.append(element)

def Pop(stack):
    if isEmpty(stack):
        print("Underflow")
        return None
    else:
        element = stack.pop()
        return element

def size(stack):
    return len(stack)

def peek(stack):
    if isEmpty(stack):
        print("Stack is empty")
        return None
    else:
        element = stack[len(stack)-1]
        return element

def display(stack):
    if isEmpty(stack):
        print("Stack is empty")
    else:
        print("The element(s) of the stack is:")
        for i in range(len(stack)-1,-1,-1):
            print(stack[i])

print("----------------------")
print("To push enter 1:")
print("To pop enter 2:")
print("To peek enter 3:")
print("To display enter 4:")
print("To know the size of the stack enter 5:")
print("To check if stack is empty or not enter 6:")
print("Enter 7 to exit:")
print("----------------------")
while True:
    opt = int(input("Enter your choice"))
    if opt == 1:
        element = int(input("Enter element to push:"))
        push(element,stack)
    elif opt == 2:
        print("The poped item is:",Pop(stack))
    elif opt == 3:
        print("The top item is:",peek(stack))
    elif opt == 4:
        display(stack)
    elif opt == 5:
        print("The size of the stack is:",size(stack))
    elif opt == 6:
        ot = isEmpty(stack)
        if ot == True:
            print("Stack is empty")
        else:
            print("Stack is not empty")
    elif opt == 7:
        break
    else:
        print("Invalid input")
```
